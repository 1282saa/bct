#### 2.3.1 시간 기반 라벨링

**왜 D-3m/6m/9m/12m로 라벨링했는가?**

조기경보 시스템은 "폐업 전" 신호를 포착해야 한다. 시간 기반 라벨링은 다음 기준으로 설계하였다:

```python
months_to_close = (폐업월 - 관측월) * 12 + (폐업일.month - 관측월.month)

if months_to_close <= 3:  상태 = 'D-3m'   # 3개월 이내 폐업 예정
elif months_to_close <= 6:  상태 = 'D-6m'   # 6개월 이내
elif months_to_close <= 9:  상태 = 'D-9m'   # 9개월 이내
elif months_to_close <= 12: 상태 = 'D-12m'  # 12개월 이내
else: 상태 = '정상영업'
```

**근거**:
- 3개월: 즉각 개입 필요 (긴급)
- 6개월: 집중 모니터링 (고위험)
- 9~12개월: 예방적 지원 (중위험)

#### 2.3.2 K-means 클러스터링

| 6 | 0.4621 | 1.1456 | 과분할 |

**Silhouette Score 해석**:
- 0.7~1.0: 강력한 구조

#### 2.3.3 통계 검정 (T-test)

**왜 T-검정과 Cohen's d를 함께 사용했는가?**

1. **T-검정**: 통계적 유의성 (p-value) → "차이가 우연인가?"
2. **Cohen's d**: 효과 크기 (Effect Size) → "차이가 실질적으로 큰가?"

대규모 데이터(86,263개)에서는 작은 차이도 통계적으로 유의하게 나타날 수 있으므로, **실질적 효과 크기**를 함께 검토해야 한다.

**Cohen's d 해석 기준**:

| \|d\| | 효과 크기 | 의미 |
|------|----------|------|
| < 0.2 | 무시 가능 | 통계적으로만 유의, 실무 의미 없음 |
| 0.2~0.5 | 작음 | 실무적으로 의미 있는 차이 |
| 0.5~0.8 | 중간 | 명확한 차이 |
| ≥0.8 | 큼 | 매우 큰 차이 |

**결과 요약**:

| 지표군 | 유의변수 수 (p<0.05) | 평균 \|Cohen's d\| | 최대 \|Cohen's d\| | 해석 |
|--------|---------------------|-------------------|-------------------|------|
| **고객지표** | 11개 | **0.163** | **0.324** | 효과 크기 가장 큼 |
| **운영지표** | 18개 | 0.124 | 0.226 | 작은 효과 |
| 접근성지표 | 5개 | 0.074 | 0.147 | 효과 매우 작음 |

| 경쟁밀집지표 | 3개 | 0.058 | 0.068 | 효과 매우 작음 |

**총 48개 변수에서 통계적으로 유의한 차이 발견** (p < 0.05)

**효과 크기가 작은~중간인 변수 (\|Cohen's d\| ≥ 0.2)**:

| 변수 | Cohen's d | 정상 평균 | 고위험 평균 | 올바른 해석 |
|------|-----------|----------|------------|------------|
| **남성 60대이상 고객 비중** | **-0.324** | 9.05% | 5.74% | 정상 그룹이 3.3%p 더 높음 |

**중요한 발견**:

1. **통계적 유의성**: 48개 변수에서 p < 0.05로 우연이 아님을 확인

#### 2.3.4 SHAP 분석

**왜 SHAP을 사용했는가?**

전통적 통계 기법(T-test)은 단변량 차이만 검증한다. 그러나 실제 폐업은 변수 간 상호작용의 결과다. SHAP(SHapley Additive exPlanations)은 게임 이론 기반으로 각 변수의 한계 기여도를 정량화한다.

**SHAP 장점**:
1. **비선형 관계 포착**: XGBoost와 결합하여 복잡한 패턴 학습
2. **상호작용 고려**: 변수 간 시너지 효과 반영
3. **방향성 제공**: 증가 시 폐업 확률 상승/하락 파악

**분석 설정**:

#### 2.3.5 LISA 공간분석

**왜 LISA 분석을 추가했는가?**

폐업은 **공간적으로 독립적이지 않다**. 인근 가맹점의 폐업이 연쇄 폐업을 유발할 수 있다 (공간 자기상관, Spatial Autocorrelation).

LISA(Local Indicators of Spatial Association)는 **지역별 공간 군집** 패턴을 탐지한다.

**방법**:
